# Simple Text in Visualisations

```python
import os.path

from matplotlib import pyplot as plt
import matplotlib as mpl
```

```python
# makes the inline-figures in notebooks look crisper
%config InlineBackend.figure_format = 'svg'
```

## Example Text
Based on an example in Cole Nussbaumer Knaflic (2015): Storytelling with Data.

```python
# create new figure
fig, ax = plt.subplots(figsize=(7.45, 4.9)) # width, height in inches

# remove chart border
for spine in ['top', 'right', 'left', 'bottom']:
    ax.spines[spine].set_visible(False)

# change the appearance of ticks, tick labels, and gridlines 
ax.tick_params(bottom=False, left=False, labelleft=False, labelbottom=False)

# add text to the plot
ax.text(0.0, 0.5,
         'Based on survey of 100 students conducted\n'
         'before and after pilot program\n'
         '(100% response rate on both surveys).\n'
         '68% of kids expressed interest towards science,\n'
         'compared to 44% going into the program.',
         fontsize=19,
         color='black')

plt.show()
```

## Simple Bar Chart

```python
fig, ax = plt.subplots(figsize=(7.45, 4.9))

# remove chart border
for spine in ['top', 'right', 'left']:
    ax.spines[spine].set_visible(False)

ax.tick_params(bottom=False, left=False, labelleft=False, labelbottom=True)

ax.set_title('The percentage of students with interest in science\n before and after the pilot program')
    
bars = ax.bar(['before', 'after'], [44, 68], # groups on x, their bar heights
               color='black',
               linewidth=0.5,
               width=0.55)

ax.bar_label(bars, fmt='%.0f%%', padding=3)

plt.show()
```

## Visualisation with Text Elements

```python
# create new figure
fig, ax = plt.subplots(figsize=(7.45, 4.9)) # width, height in inches

# remove chart border
for spine in ['top', 'right', 'left', 'bottom']:
    ax.spines[spine].set_visible(False)

# change the appearance of ticks, tick labels, and gridlines 
ax.tick_params(bottom=False, left=False, labelleft=False, labelbottom=False)

# title the plot
ax.text(-0.15, 1.03,
         'Pilot program was a success'+' '*27,
         fontsize=26,
         color='black'
        )

# footnote with the data source
ax.text(-0.15, 0.04,
         'Based on survey of 100 students conducted '
         'before and after pilot program '
         '(100% response rate on both surveys).',
         fontsize=10.5,
         color='black')

# add note to the plot
ax.text(-0.15, 0.81,
         'After the pilot program,',
         fontsize=19,
         color='black')

# add note to the plot
ax.text(-0.15, 0.41,
         '$\\bf{68\%}$', # use mathtext \\bf for bold text
         fontsize=122,
         color='black')

# add note to the plot
ax.text(-0.15, 0.25,
         '$\\bf{of\ kids\ expressed\ interest\ towards\ science,}$',
         fontsize=19,
         color='black')

# add note to the plot
ax.text(-0.15, 0.17,
         'compared to 44% going into the program.',
         fontsize=19,
         color='black')

plt.show()
```

## Using Shades of Grey

```python
# create new figure
fig, ax = plt.subplots(figsize=(7.45, 4.9)) # width, height in inches

# remove chart border
for spine in ['top', 'right', 'left', 'bottom']:
    ax.spines[spine].set_visible(False)

# change the appearance of ticks, tick labels, and gridlines 
ax.tick_params(bottom=False, left=False, labelleft=False, labelbottom=False)

# title the plot
ax.text(-0.15, 1.03,
         'Pilot program was a success'+' '*28,
         fontsize=26,
         color='white',
         # text color to white and
         # put a rectangular box around the text to set a background color
         bbox={'facecolor': 'darkgrey', 'pad': 10, 'edgecolor': 'none'}
        )

# footnote with the data source
ax.text(-0.15, 0.04,
         'Based on survey of 100 students conducted '
         'before and after pilot program '
         '(100% response rate on both surveys).',
         fontsize=10.5,
         color='black')

# add note to the plot
ax.text(-0.15, 0.81,
         'After the pilot program,',
         fontsize=19,
         color='black')

# add note to the plot
ax.text(-0.15, 0.41,
         '$\\bf{68\%}$', # use mathtext \\bf for bold text
         fontsize=122,
         color='black')

# add note to the plot
ax.text(-0.15, 0.25,
         '$\\bf{of\ kids\ expressed\ interest\ towards\ science,}$',
         fontsize=19,
         color='black')

# add note to the plot
ax.text(-0.15, 0.17,
         'compared to 44% going into the program.',
         fontsize=19,
         color='black')

plt.show()
```

```python
import matplotlib.pyplot as plt
import os
from dotenv import load_dotenv

load_dotenv()

work_dir = os.path.join(os.getenv('OUTPUT_FOLDER'), '03_exercise')
os.makedirs(work_dir, exist_ok=True)

# Create a compact figure
plt.figure(figsize=(6, 4))  # Reduced figure size
plt.axis('off')  # Turn off the axes

# Add the title with a bold color
plt.text(0.5, 0.9, 'Children with a "Traditional" Stay-at-Home Mother',
         fontsize=16, weight='bold', ha='center', color='darkblue')

# Add the subtitle with a softer color
plt.text(0.5, 0.82, '% of children with a married stay-at-home mother with a working husband',
         fontsize=10, ha='center', color='darkgreen')

# Add the data points with vibrant colors
plt.text(0.3, 0.6, '41%', fontsize=30, weight='bold', ha='center', color='crimson')
plt.text(0.3, 0.55, '1970', fontsize=16, ha='center', color='teal')

plt.text(0.7, 0.6, '20%', fontsize=30, weight='bold', ha='center', color='orange')
plt.text(0.7, 0.55, '2012', fontsize=16, ha='center', color='purple')

# Add a downward arrow pointing to the bottom right
plt.text(0.5, 0.58, '', fontsize=40, ha='center', color='green')  # Downward arrow

# Add the note with a neutral color
note = ('Note: Based on children younger than 18. Their mothers are categorised based on employment status in 1970 and 2012.\n'
        'Source: Pew Research Center analysis of March current Population Surveys Integrated Public Use Microdata Series (IPUMS-CPS), 1971 and 2013.')
plt.text(0.5, 0.35, note, fontsize=8, ha='center', wrap=True, color='gray')

# Show the figure
plt.tight_layout()
plt.savefig(os.path.join(work_dir, 'traditional_mothers.png'), dpi=300, bbox_inches='tight')
plt.show()
```
