# Axes Methods

```python
import matplotlib.pyplot as plt
# makes the inline-figures in notebooks look crisper
%config InlineBackend.figure_format = 'svg'
```

Bar chart:

```python
y = ["France", "Spain", "Canada", "UK", "Australia"]
width = [2, 5, 3, 4, 5]

fig, ax = plt.subplots()
ax.barh(y, width)

# Set font size for all labels and text elements
ax.set_xlabel(ax.get_xlabel(), fontsize=18)
ax.set_ylabel(ax.get_ylabel(), fontsize=18)
ax.set_title(ax.get_title(), fontsize=18)
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(18)

plt.show()
```

Area chart:

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 5, 3, 4, 5, 2, 6]

fig, ax = plt.subplots()
ax.fill_between(x, y)

# Set font size for all labels and text elements
ax.set_xlabel(ax.get_xlabel(), fontsize=18)
ax.set_ylabel(ax.get_ylabel(), fontsize=18)
ax.set_title(ax.get_title(), fontsize=18)
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(18)
plt.show()
```

Bubble chart:

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 5, 3, 4, 5, 2, 6]
s = [100, 200, 300, 400, 200, 100, 500]

fig, ax = plt.subplots()
ax.scatter(x, y, s=s)

# Set font size for all labels and text elements
ax.set_xlabel(ax.get_xlabel(), fontsize=18)
ax.set_ylabel(ax.get_ylabel(), fontsize=18)
ax.set_title(ax.get_title(), fontsize=18)
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(18)

plt.show()
```

Lollipop chart:

```python
y = [2, 5, 3, 4, 5, 2, 6]

fig, ax = plt.subplots()
ax.stem(y)

# Set font size for all labels and text elements
ax.set_xlabel(ax.get_xlabel(), fontsize=18)
ax.set_ylabel(ax.get_ylabel(), fontsize=18)
ax.set_title(ax.get_title(), fontsize=18)
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(18)

plt.show()
```

Pie chart:

```python
x = [6, 2, 3, 4, 5]
labels = ["matplotlib", "ggplot2", "plotnine", "D3.js", "PowerBI"]

fig, ax = plt.subplots()
ax.pie(x, labels=labels, textprops={'fontsize': 18})

plt.show()
```
