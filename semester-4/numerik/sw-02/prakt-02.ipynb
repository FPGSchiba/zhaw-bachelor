## Aufgabe 4

### a) & b)

```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: 1 - np.cos(x)
x = np.linspace(-0.5e-7, 0.5e-7, 1000)
y = f(x)

plt.figure()
plt.plot(x, y)
plt.show()
```

### c)

Man kann sehr schön auslöschungen beobachten im Bereich von 0.5e-7. Das liegt daran, dass die Werte sehr nah aneinander liegen und die Genauigkeit der Gleitkommazahlen nicht ausreicht um die Werte korrekt zu berechnen.

## Aufgabe 6

### a)

```python
f_poly = lambda x: x**6 - 12*x**5 + 60*x**4 - 160*x**3 + 240*x**2 - 192*x + 64
x = np.linspace(1.99, 2.01, 400)
y = f_poly(x)

plt.figure()
plt.plot(x, y)
plt.show()
```

### b)

Man sieht hier ziemlich schön die Rundungsfehler, die geschehen bei den ganzen Operationen. Die Werte sind sehr nah aneinander und die Genauigkeit der Gleitkommazahlen reicht nicht aus um die Werte korrekt zu berechnen.
