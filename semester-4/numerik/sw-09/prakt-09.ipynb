## Aufgabe 2

### b)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([-1, 0, 2, 3])
y = np.array([-1, -3, 1, -27])
A = np.vander(x, 4)
p = np.linalg.solve(A, y)

xx = np.linspace(-2, 4, 100 )
yy = np.polyval(p, xx)
plt.plot( x, y, "b*", xx, yy, "r" )
plt.title( "Interpolationspolynom" )
plt.show()
```

## Aufgabe 3

```python
l0 = lambda x: (x*(x-2)*(x-4))/(-2*-4*-6)
l1 = lambda x: ((x+2)*(x-2)*(x-4))/(-2*2*-4)
l2 = lambda x: ((x+2)*x*(x-4))/(4*2*-2)
l3 = lambda x: ((x+2)*(x-2)*x)/(6*4*2)

x = np.array([-2, 0, 2, 4])
y = np.array([-31, -5, 13, 119])

p = lambda x: y[0]*l0(x) + y[1]*l1(x) + y[2]*l2(x) + y[3]*l3(x)

x_1 = 1
y_1 = p(x_1)

print(y_1)

xx = np.linspace(-2, 4, 100)
yy = p(xx)
plt.plot(x, y, "b*", xx, yy, "r", x_1, y_1, "go")
plt.title( "Interpolationspolynom" )
plt.show()
```

## Aufgabe 5

```python
x = np.array([23, 35, 48, 61])
y = np.array([320, 490, 540, 500])

l0 = lambda x: ((x-35)*(x-48)*(x-61))/((23-35)*(23-48)*(23-61))
l1 = lambda x: ((x-23)*(x-48)*(x-61))/((35-23)*(35-48)*(35-61))
l2 = lambda x: ((x-23)*(x-35)*(x-61))/((48-23)*(48-35)*(48-61))
l3 = lambda x: ((x-23)*(x-35)*(x-48))/((61-23)*(61-35)*(61-48))

p = lambda x: y[0]*l0(x) + y[1]*l1(x) + y[2]*l2(x) + y[3]*l3(x)

x_1 = 40
y_1 = p(x_1)

print(y_1)

xx = np.linspace(20, 65, 100)
yy = p(xx)
plt.plot(x, y, "b*", xx, yy, "r", x_1, y_1, "go")
plt.title( "Interpolationspolynom" )
plt.show()
```

## Aufgabe 6

### a)

```python
x = np.array([1981, 1984, 1989, 1993, 1997, 2000, 2001, 2003, 2004, 2010]) - 1981
y = np.array([0.5, 8.2, 15, 22.9, 36.6, 51, 56.3, 61.8, 65, 76.7])

p_1 = np.polyfit(x, y, 9)
xx = np.linspace(0, 29, 100)
yy = np.polyval(p_1, xx)

plt.plot(x + 1981, y, "b*", xx + 1981, yy, "r")
plt.title( "Interpolationspolynom" )
plt.show()
```

### b)

```python
import scipy

p_2 = scipy.interpolate.interp1d(x, y, kind='cubic')

xx = np.linspace(0, 29, 100)
yy = p_2(xx)
plt.plot(x + 1981, y, "b*", xx + 1981, yy, "r")
plt.title( "Interpolationspolynom" )
plt.show()
```

### c)

```python
xx = np.linspace(0, 29, 100)
yy = p_2(xx)
yyy = np.polyval(p_1, xx)

plt.plot(x + 1981, y, "b*", xx + 1981, yy, "r", xx + 1981, yyy, "g")
plt.title( "Interpolationspolynom" )
plt.legend(['Daten', 'Spline', 'Polynom'])
plt.show()
```
