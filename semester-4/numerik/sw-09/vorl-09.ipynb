```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2, 5)
y = np.array([1, 1, 0, 0, 3])
A = np.vander(np.array([0, 0.5, 1, 1.5, 2]), 5)
p = np.linalg.solve( A, y )

xx = np.linspace(-1, 3, 100 )
yy = np.polyval( p, xx )
plt.plot( x, y, "b*", xx, yy, "r" )
plt.title( "Interpolationspolynom" )
plt.show()
```

```python
x = np.linspace(0, 2, 5)
y = np.array([1, 1, 1, 1, 1])
A = np.vander(np.array([0, 0.5, 1, 1.5, 2]), 5)
p = np.linalg.solve( A, y )

print(p)

xx = np.linspace(-1, 3, 100 )
yy = np.polyval( p, xx )
plt.plot( x, y, "b*", xx, yy, "r" )
plt.title( "Interpolationspolynom" )
plt.show()
```
