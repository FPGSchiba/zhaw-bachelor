```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Funktion fÃ¼r den Multidimensionalen Newton

def newton(f, J, x0, tol=1.e-10):
    max_loops = 1000
    x = x0
    k = 0
    res = np.linalg.norm(f(x))
    while res > tol and k < max_loops:
        k += 1
        d = np.linalg.solve(J(x), f(x))
        x = x - d
        res = np.linalg.norm(f(x))
    return x, k
```

```python
# Definition der Funktion
f = lambda x: (np.exp(x[0]**2 + x[1]**2)-3, np.sin(3*x[0]+3*x[1])-x[0]-x[1])

# Definition der Jacobi-Matrix
J = lambda x: np.array([[2*x[0]*np.exp(x[0]**2 + x[1]**2), 2*x[1]*np.exp(x[0]**2 + x[1]**2)],
                        [3*np.cos(3*x[0]+3*x[1]) - 1, 3*np.cos(3*x[0]+3*x[1]) - 1]])
```

```python
x_lin = np.linspace(-2, 2, 100)
y_lin = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x_lin, y_lin)

# Create figure
fig, ax = plt.subplots()

# Calculate Z values for both functions
Z1 = np.array([[f([x, y])[0] for x in x_lin] for y in y_lin])
Z2 = np.array([[f([x, y])[1] for x in x_lin] for y in y_lin])

# Plot both contours
ax.contour(X, Y, Z1, levels=[0], colors='r')
ax.contour(X, Y, Z2, levels=[0], colors='b')

# Add labels and legend
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.grid(True)
plt.show()
```

```python
# Erste Nullstelle
result = newton(f, J, (0.25, -1), tol=1.e-6)
print(result)

# Auswertung - LÃ¤nge der LÃ¶sung
print(abs(np.linalg.norm(result[0]) - np.sqrt(np.log(3))))

# Zweite Nullstelle
result = newton(f, J, (0.75, -0.5), tol=1.e-6)
print(result)

# Auswertung - LÃ¤nge der LÃ¶sung
print(abs(np.linalg.norm(result[0]) - np.sqrt(np.log(3))))

# Dritte Nullstelle
result = newton(f, J, (1, -0.25), tol=1.e-6)
print(result)

# Auswertung - LÃ¤nge der LÃ¶sung
print(abs(np.linalg.norm(result[0]) - np.sqrt(np.log(3))))

# Vierte Nullstelle
result = newton(f, J, (-0.25, 1), tol=1.e-6)
print(result)

# Auswertung - LÃ¤nge der LÃ¶sung
print(abs(np.linalg.norm(result[0]) - np.sqrt(np.log(3))))

# FÃ¼nfte Nullstelle
result = newton(f, J, (-0.75, 0.75), tol=1.e-6)
print(result)

# Auswertung - LÃ¤nge der LÃ¶sung
print(abs(np.linalg.norm(result[0]) - np.sqrt(np.log(3))))

# Sechste Nullstelle
result = newton(f, J, (-1, 0.25), tol=1.e-6)
print(result)

# Auswertung - LÃ¤nge der LÃ¶sung
print(abs(np.linalg.norm(result[0]) - np.sqrt(np.log(3))))
```
