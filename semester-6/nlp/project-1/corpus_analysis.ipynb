```python
from workbench import *
import matplotlib.pyplot as plt
import csv
```

```python
texts = []
with open('data/test.csv', 'r', encoding='utf8') as f:
    for row in csv.reader(f):
        texts.append(row[1])

print(len(texts))
```

```python
unique_characters = get_number_character_set(texts)
print(len(unique_characters))
```

```python
get_number_of_characters(texts)
```

```python
get_number_of_words(texts)
```

```python
get_number_of_sentences(texts)
```

```python
get_number_of_different_words(texts)
```

```python
num_words = []
num_sentences = []
for text in texts:
    num_words.append(get_number_of_words(text))
    num_sentences.append(get_number_of_sentences(text))
```

```python
plt.hist(num_words, bins=20)
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.title('Distribution of Number of Words in Texts')
plt.show()
```

```python
plt.hist(num_sentences, bins=20)
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.title('Distribution of Number of Words in Texts')
plt.show()
```
